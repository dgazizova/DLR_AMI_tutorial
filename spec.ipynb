```python
import matplotlib.pyplot as plt
from functions import *
import numpy as np
```

Self energy:
$$
\Sigma(i\omega) = \iiint A(\omega_1)A(\omega_2) A(\omega_3) \frac{[f(\omega_{1}) - f(\omega_{2})] [n(\omega_{2} - \omega_{1}) + f(-\omega_{3}) ]}{i\omega + \omega_{1} - \omega_{2} - \omega_{3}} d\omega_1 d\omega_2 d\omega_3
$$


```python
beta = 20
w = np.linspace(-5, 5, 100)
iw = matsubara("F", n=np.arange(-20, 20), beta=beta)
SE = SelfEnergy(beta=beta)
SE_spec_rf = SE.SE_full_spectral(2000000, 2, w+0.1*1j) # small Gamma part 0.1
SE_spec_iw = SE.SE_full_spectral(2000000, 2, iw*1j)
```

Bethe lattice spectral:
$$
A(w) = \frac{1}{2\pi t^2} \sqrt{4t^2 - w^2}
$$
$t = 1$

```python
A = bethe_lattice_dos(w, t=1)
plt.plot(w, A,  marker=".", label="Bethe lattice spectral")
plt.ylabel("A")
plt.xlabel(r"$\omega$")
plt.legend()
plt.show()
```

```python
fig = plt.figure()
plt.subplot(2, 1, 1)
plt.plot(iw, np.real(SE_spec_iw), marker=".", label="Full spectral")
plt.ylabel(r"$\Sigma(i\omega)$")
plt.xlabel(r"$i\omega$")
plt.legend()
plt.subplot(2, 1, 2)
plt.plot(iw, np.imag(SE_spec_iw), marker=".", label="Full spectral")
plt.ylabel(r"$\Sigma(i\omega)$")
plt.xlabel(r"$i\omega$")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# plot SE rf
plt.subplot(2, 1, 1)
plt.ylabel(r"$\Sigma(\omega)$")
plt.xlabel(r"$\omega$")
plt.plot(w, np.real(SE_spec_rf), marker=".", label="Full spectral")
plt.legend()
plt.subplot(2, 1, 2)
plt.ylabel(r"$\Sigma(\omega)$")
plt.xlabel(r"$\omega$")
plt.plot(w, np.imag(SE_spec_rf), marker=".", label="Full spectral")
plt.legend()
plt.tight_layout()
plt.show()
```
